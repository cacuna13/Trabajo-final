```python
gf1 = basedatos.groupby(['Tiempo']).mean()
gf1 = gf1.iloc[:,:20]
gf1
```

```python
eje_x = ("2010", "2011", "2012","2013", "2014", "2015", "2016", "2017", "2018", "2019")
eje_y = gf1['Total'].tolist()


#crear la grafica
plt.rcdefaults()
fig, ax = plt.subplots()

#este se alimenta del la lista denominada eje:x
y_pos = np.arange(len(eje_x))

ax.barh(y_pos, eje_y, align='center', color="#006400", alpha=0.5)

ax.set_yticks(y_pos, labels=eje_x, color = "black")
ax.invert_yaxis() #labels read top-to-bottom
ax.set_xlabel('Consumo de Combustible').set_color('#483d8b')
ax.set_title('Todos los Combustibles')

plt.show()
```

```python
gf2 = basedatos.groupby(['Tiempo']).mean()
gf2 = gf2.iloc[:,:20]
gf2
```

```python
fig, ax = plt.subplots(figsize=(10,4), subplot_kw=dict(aspect="equal"))

data = gf2['Total'].tolist()
estilo = ['2010','2011','2012','2013','2014','2015','2016','2017', '2018', '2019']

def func(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return "{:.1f}%".format(pct, absolute)
wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))
ax.legend(wedges, estilo,
          title="Estilo",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 0.9))

plt.setp(autotexts, size=6, weight="bold")

ax.set_title("Consumos anuales Todos los Combustibles")

plt.show()
```


<p Style="text-aling: justify" >Como podemos observar en los gráficos si bien es cierto se ha visto un des aceleramiento en el consumo de combustibles aun continua el consumo a la alza por lo que los esfuerzos realizados al 2019 no son suficientes.</p>
